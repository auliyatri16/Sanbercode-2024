Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
#code here
fig, ax = plt.subplots()

plt.plot (x, cos, color='red', label='Cosx')
plt.plot (x, cos3, linestyle='dashed', color='green', label='3 Cosx')
plt.plot (x, sin, linestyle='dashdot', color='blue', label='Sin(x-90)')
plt.plot (x, sin2, linestyle='', marker= 'x', color='yellow', label='2 Sinx')
ax.set_xlabel('X-Axis Label')
ax.set_ylabel('Y-Axis Label')
ax.set_title('Grafik Cos Sin')
plt.legend(loc='center', fontsize='xx-small')


plt.show()
```
