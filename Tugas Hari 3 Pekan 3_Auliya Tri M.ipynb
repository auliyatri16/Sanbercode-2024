<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(2*-np.pi, 2*np.pi, 200)

tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
#code here
fig, ax= plt.subplots(nrows=1, ncols=3, sharex=True, sharey=True, figsize=(9,4))
ax[0].plot(tan, color='blue', label='tanX')
ax[1].plot(cos, color='red', label='sinX')
ax[2].plot(sin, color='green', label='cosX')

ax[0].set_title('Grafik tanx')
ax[1].set_title('Grafik sinx')
ax[2].set_title('Grafik cosx')

fig.legend(bbox_to_anchor=(0.5, 1.1), loc='upper center', fontsize='xx-small')


plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
#code here
fig, ax= plt.subplots()
ax.plot(x,y)
ax.set_title('Grafik Cosx')

ins1 = ax.inset_axes([-0.42, -0.1, 0.6, 0.4])
ins1.plot(x,y2, color='red')
ins1.plot(x,y, color='blue')
ins1.set_title('Grafik CosX^2', fontsize='x-small')

ins2 = ax.inset_axes([0.93, -0.1, 0.6, 0.4])
ins2.plot(x,y3, color='green')
ins2.plot(x,y, color='blue')
ins2.set_title('Grafik CosX^3', fontsize='x-small')

ins3 = ax.inset_axes([0.93, 0.9, 0.6, 0.4])
ins3.plot(x,y4, color='black')
ins3.plot(x,y, color='blue')
ins3.set_title('Grafik CosX^4', fontsize='x-small')

ins4 = ax.inset_axes([-0.42, 0.9, 0.6, 0.4])
ins4.plot(x,y5, color='yellow')
ins4.plot(x,y, color='blue')
ins4.set_title('Grafik CosX^5', fontsize='x-small')


plt.show()
```
